## Price by Ingredients Aggregation

```python
inFile = open("bfast.json",'r')
meals = json.load(inFile)["results"]
```

```python
def price_breakdown(rid):
    url = f"https://api.spoonacular.com/recipes/{rid}/priceBreakdownWidget.json?apiKey={api_key}"
    prices = json.loads(requests.get(url).text)["ingredients"]
    return prices
```

```python
price_breakdown(715497)
```

```python
ingredients_json = {}
cache_breakdowns = {}
```

```python
iter = 0
for meal in meals:
    if meal["id"] not in cache_breakdowns:
        breakdown = price_breakdown(meal["id"])
        cache_breakdowns[meal["id"]] = breakdown
    breakdown = cache_breakdowns[meal["id"]]
    try:
        assert(len(breakdown) == len(meal["nutrition"]["ingredients"]))
    except:
        for i in range(len(breakdown)):
            print(breakdown[i]["name"], end = ", ")
        print()
        for i in range(len(meal["nutrition"]["ingredients"])):
            print(meal["nutrition"]["ingredients"][i]["name"], end = ", ")
        print()
        print()
        # print(breakdown, len(breakdown))
        # print(meal["nutrition"]["ingredients"], len(meal["nutrition"]["ingredients"]))
        continue
    iter += 1
    for i, ingredient in enumerate(meal["nutrition"]["ingredients"]):
        if ingredient["id"] in ingredients_json:
            try:
                assert(ingredients_json[ingredient["id"]]["name"] == breakdown[i]["name"])
            except:
                for i in range(len(breakdown)):
                    print(breakdown[i]["name"], end = ", ")
                print()
                for i in range(len(meal["nutrition"]["ingredients"])):
                    print(meal["nutrition"]["ingredients"][i]["name"], end = ", ")
                print()
                print()
            continue
        ingredients_json[ingredient["id"]] = {"name": breakdown[i]["name"], "metric_price": breakdown[i]["price"] / breakdown[i]["amount"]["metric"]["value"], "metric_unit": breakdown[i]["amount"]["metric"]["unit"], "us_price": breakdown[i]["price"] / breakdown[i]["amount"]["us"]["value"], "us_unit": breakdown[i]["amount"]["us"]["unit"]}
```
